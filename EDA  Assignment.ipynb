```python

def univar_cat(col,ylog=False,label_rotation=False,horizontal_layout=True):
    t = df[col].value_counts()
    dff = pd.DataFrame({col: t.index,'Number of contracts': t.values})

    # Calculate the % of target=1 for particular category
    cat_perc = df[[col, 'TARGET']].groupby([col],as_index=False).mean()
    cat_perc["TARGET"] = cat_perc["TARGET"]*100
    cat_perc.sort_values(by='TARGET', ascending=False, inplace=True)
    
    sns.barplot(x = col, 
                    y='TARGET', 
                    order=cat_perc[col], 
                    data=cat_perc,
                    palette='Set1')
    
    if(label_rotation):
        plt.xticks(rotation=90)
        plt.xlabel("")
    plt.ylabel('Percent of Defaulters [%]', fontsize=15,fontweight=3, color='Blue')
    plt.tick_params(axis='both', which='major', labelsize=15)
    plt.title(col + " Defaulter %", fontdict={'fontsize' : 20, 'fontweight' : 20, 'color' : 'Blue'}) 
    plt.xlabel("")
    plt.show();
```
